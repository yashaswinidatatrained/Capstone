```python
import selenium
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException

from bs4 import BeautifulSoup
import requests

import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

### Scraping For Austin, Texas

```python
#opening the rentberry Austin,Tx page on automated chrome browser
driver = webdriver.Chrome()
driver.get('https://rentberry.com/apartments/s/las-vegas-nv')
time.sleep(2)
```

```python
rent_URL = []
rent_Price = []
```

```python
# to scrap URL of 66 pages
start = 0
end = 40

for page in range(start,end):
    url_tags= driver.find_elements(By.XPATH,'//a[@class="property-card-bottom-address__text link-ghost font-xs one-line-string"]')
    price = driver.find_elements(By.XPATH,'//rb-property-card-bottom[@class="property-card__bottom"]/div/div/rb-property-card-price/div/p')
   
    for _ in range(100):
        driver.execute_script("window.scrollBy(0,1000)")
        
    for t in url_tags:
        rent_URL.append(t.get_attribute('href'))
    for a in price:
        rent_Price.append(a.text)
        
    next_page = driver.find_element(By.XPATH,'//span[@class="pagination-arrow-next"]')
    next_page.click()
 
    time.sleep(5)
```

```python
rent_Pets = []
rent_Area = []
rent_Address = []
rent_Deposit = []
rent_Beds = []
rent_Bathroom = []
rent_Washer_Dryer =[]
rent_AC = []
rent_Parking=[]
rent_Fireplace=[]
rent_Dishwasher=[]
rent_Floors=[]
rent_Roofdeck=[]
rent_Storage=[]
```

```python
for t in rent_URL[444:]:
    driver.get(t)
    time.sleep(5)
    
    
    try:
        add = driver.find_element(By.XPATH,'/html/body/rb-core/div/rb-property/div/rb-property-main-info/section/div/div/rb-property-data/div/h4/span/span')
        rent_Address.append(add.text)
    except NoSuchElementException:
        rent_Address.append(np.nan)
        
    try:
        bed = driver.find_element(By.XPATH,"//*[contains(@class,'bedrooms')]")
        rent_Beds.append(bed.text)
    except NoSuchElementException:
        rent_Beds.append(np.nan) 
        
    try:
        bath = driver.find_element(By.XPATH,"//*[contains(@class,'bathrooms')]")
        rent_Bathroom.append(bath.text)
    except NoSuchElementException:
        rent_Bathroom.append(np.nan) 
        
    try:
        sqft = driver.find_element(By.XPATH,"//*[contains(@class,'space')]")
        rent_Area.append(sqft.text)
    except NoSuchElementException:
        rent_Area.append(np.nan)
        
    try:
        pets = driver.find_element(By.XPATH,"//*[contains(@class,'dog-icon facilities-param-icon')]" or "//*[contains(@class,'cat-icon facilities-param-icon')]")
        rent_Pets.append('Yes')
    except NoSuchElementException:
        rent_Pets.append('No')
        
    try:
        adv = driver.find_element(By.XPATH,'/html/body/rb-core/div/rb-property/div/div[2]/div/rb-apply-block/div/div/rb-button-square/button/span/span[2]/span[1]')
        rent_Deposit.append(adv.text.replace("Apply for $",''))
    except NoSuchElementException:
        rent_Deposit.append(np.nan)
        
    try:
        was = driver.find_element(By.XPATH, "//*[contains(@class,'icon washerUnit')]")
        rent_Washer_Dryer.append('Yes')
    except NoSuchElementException:
        rent_Washer_Dryer.append('No')   
        
    try:
        ac = driver.find_element(By.XPATH,"//*[contains(@class,'airConditioning icon')]")
        rent_AC.append('Yes')
    except NoSuchElementException:
        rent_AC.append('No')
       
    try:
        park = driver.find_element(By.XPATH,"//*[contains(@class,'icon parking')]")
        rent_Parking.append('Yes')
    except NoSuchElementException:
        rent_Parking.append('No') 

    try:
        dish = driver.find_element(By.XPATH,"//*[contains(@class,'dishwasher icon')]")
        rent_Dishwasher.append('Yes')
    except NoSuchElementException: 
        rent_Dishwasher.append('No')
    try:
        hard = driver.find_element(By.XPATH,"//*[contains(@class,'hardwoodFloors icon')]")
        rent_Floors.append('Yes')
    except NoSuchElementException:
        rent_Floors.append('No')
        
    try:
        roof = driver.find_element(By.XPATH,"//*[contains(@class,'icon roofDeck')]")
        rent_Roofdeck.append('Yes')  
    except NoSuchElementException:
        rent_Roofdeck.append('No')
        
    try:
        store = driver.find_element(By.XPATH, "//*[contains(@class,'icon storage')]")
        rent_Storage.append('Yes')
    except NoSuchElementException: 
        rent_Storage.append('No')
                    
    try:
        fire = driver.find_element(By.XPATH, "//*[contains(@class,'fireplace icon')]")
        rent_Fireplace.append('Yes')
    except NoSuchElementException:  
        rent_Fireplace.append('No')
```
